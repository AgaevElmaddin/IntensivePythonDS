## Imports

```python
import pandas as pd
import sqlite3
```

```python
connection = sqlite3.connect('../data/checking-logs.sqlite.sqlite')
connection
```

## analyze only the users and not the admins


```python
query = '''
SELECT uid, datetime
FROM pageviews
WHERE uid LIKE 'user_%'
'''
views = pd.io.sql.read_sql(query, connection, parse_dates=['datetime'])
views
```

```python
query = '''
SELECT uid, timestamp
FROM checker
WHERE uid LIKE 'user_%'
'''
commits = pd.io.sql.read_sql(query, connection, parse_dates=['timestamp'])
commits
```

## analyze only the dates when there were both views and checker commits

```python
new_views = views['datetime'].dt.floor('d').value_counts()
new_views = new_views.rename_axis('date')
new_views = new_views.reset_index(name='views')
new_views = new_views.set_index('date')
new_views
```

```python
new_commits = commits['timestamp'].dt.floor('d').value_counts()
new_commits = new_commits.rename_axis('date')
new_commits = new_commits.reset_index(name='commits')
new_commits = new_commits.set_index('date')
new_commits
```

```python
views_commits = pd.concat([new_views, new_commits], axis=1)
views_commits = views_commits.dropna()
views_commits
```


## use size of the font should be 8
## the size of the figure is (15,8)

```python
axes = views_commits.plot(figsize=(15,8), title='Views and commits per day', color=['red', 'cyan'],
                          style=['-.', '--'], rot=90, fontsize=8)
axes.grid(axis='x')
axes.get_lines()[0].set_marker('o')
axes.get_legend().get_lines()[0].set_marker('o')
axes.set_facecolor('#F5F5F5')
```
